```python
from sklearn.datasets import make_classification, make_regression
import metric_learn as mkl
import matplotlib.pyplot as plt
from sklearn.manifold import TSNE
import numpy as np
```

```python
def plot_tsne(X, y, name, colormap=plt.cm.Paired):
    plt.figure(figsize=(8, 6))

    # clean the figure
    plt.clf()

    tsne = TSNE()
    X_embedded = tsne.fit_transform(X)
    plt.scatter(X_embedded[:, 0], X_embedded[:, 1], c=y, cmap=colormap)
    
    for i, txt in enumerate(name):
        plt.annotate(txt, (X_embedded[i, 0], X_embedded[i, 1]))
    
    plt.xticks(())
    plt.yticks(())

    plt.show()
```

```python
n_samples=50
X_reg, y_reg = make_regression(n_samples=n_samples, n_informative=3, n_features=5,
                               shuffle=True)
plot_tsne(X_reg, y_reg, range(n_samples), plt.cm.Oranges)
```

```python
mlkr = mkl.MLKR(n_components=2, init="auto")
X_mlkr = mlkr.fit_transform(X_reg, y_reg)
plot_tsne(X_mlkr, y_reg, range(n_samples), plt.cm.Oranges)
print (X_mlkr)
print (np.sum(np.abs(X_mlkr[0]-X_mlkr[1])))
```

```python
score_pairs = mlkr.score_pairs([
    [[3.5, 3.6, 1.0, 2.4, 3.6], [5.6, 2.4, 1.0, 2.4, 3.6]],
    [[4.5, 3.6, 1.0, 2.4, 3.6], [5.6, 2.4, 1.0, 2.4, 3.6]],
    [[8.5, 3.6, 1.0, 2.4, 3.6], [5.6, 2.4, 1.0, 2.4, 3.6]],
    [X_reg[0], X_reg[1]]

])
print (score_pairs)
```
